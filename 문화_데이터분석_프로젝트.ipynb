# 2. 국내지역별 관광명소 데이터

## 1. 데이터 불러오기

```python
import pandas as pd
import matplotlib.pyplot as plt
data1=pd.read_csv('./국내지역별 관광명소 데이터.csv')
data1.head()
```

## 2. 데이터 전처리

```python
data2=data1.drop(['대분류명',"중분류명","지점명","보조명","분류코드","지번고유코드","번지번호","법정동코드","행정동코드","도로명주소코드",
                  "위치경도","위치위도","격자코드","최종변경일자","출처명","파일명","기준일자"],axis=1)
data2.head()
```

## 3. 데이터분석

```python
data2.count()  # data23의 각 열의 개수
```

```python
data2["POI명"].unique()  # POI명에서 가진 단어들
```

```python
data3=data2["분류명"].unique()  # 분류명에서 가진 단어들
data3
```

### 1. 관광명소 시각화

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 한글 폰트 설정
plt.rc('font', family='NanumBarunGothic')  # 한글이 깨지는 것을 방지하고자 코드를 넣었습니다.

counts = data2["분류명"].value_counts()
max_value = counts.max()
norm_counts = counts / max_value
colors = cm.viridis(norm_counts)
counts.plot(kind="barh", color=colors)

# 그래프 제목과 축 레이블 설정
plt.title('국내지역 관광명소', fontsize=20)
plt.xlabel('값', fontsize=18)
plt.ylabel('관광명소', fontsize=18)

# 막대에 레이블 추가
for i, count in enumerate(counts):
    plt.text(count, i, str(count), ha='left', va='center')

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 한글 폰트 설정
plt.rc('font', family='NanumBarunGothic')  # 한글이 깨지는 것을 방지하고자 코드를 넣었습니다.

# 그래프 크기 설정
plt.figure(figsize=(10, 5))  # x축을 10 ,y축을 5로 지정하였습니다.

# 데이터 시각화
counts = 세종특별자치시["법정동명"].value_counts()
colors = cm.viridis(counts / counts.max())  # 색상 맵을 적용하여 값의 비율에 따라 색상 설정합니다.
counts.plot(kind="barh", color=colors)

counts = 세종특별자치시["법정동명"].value_counts()
colors = cm.viridis(counts / counts.max())
counts.plot(kind="barh", color=colors)

# 그래프 제목과 축 레이블 설정
plt.title('세종특별자치시 지역', fontsize=20)
plt.xlabel('값', fontsize=18)
plt.ylabel('법정동명', fontsize=18)

# 막대에 레이블 추가
for i, count in enumerate(counts):
    plt.text(count, i, str(count), ha='left', va='center')

# 그래프 출력
plt.show()
```
